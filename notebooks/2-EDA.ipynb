# Exploratory Data Analysis (EDA)

The objective of this notebook is to explore the cleaned dataset, identify meaningful patterns, and uncover business insights that support data-driven decision making.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("../data/cleaned-sample.csv")
```

## Sales by Category

```python
sales_category=df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(sales_category)
```

```python
plt.Figure((12,6))
plt.title("Total Sales by Category")
sales_category.plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Total sales($)")
plt.show()
```

### Observation

The **Technology** category generates the highest sales, with approximately **100K more in sales revenue** compared to other categories. This indicates that technology products are a major revenue driver for the business, while **Furniture** represents the second-highest sales category.

## Sales Trends Overtime

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```python
df["Year"]=df["Order Date"].dt.year
df["Month"]=df["Order Date"].dt.month_name()
```

```python
yearly=df.groupby("Year")["Sales"].sum()
months_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

monthly = df.groupby("Month")["Sales"].sum().reindex(months_order)
print(yearly)
print(monthly)
```

```python
plt.figure(figsize=(12,6))
yearly.plot(kind="line", marker="o")
plt.title("Yearly Sales Trend")
plt.xlabel("Years")
plt.ylabel("Total Sales($)")
plt.show()
```

### Observation 

The yearly sales trend shows a temporary decline between 2015 and 2016. Afterward, sales experienced steady growth and reached their highest point in 2018, generating approximately $700K in total revenue.

```python
plt.figure(figsize=(12,6))
monthly.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Total Sales($)")
plt.xticks(range(12), monthly.index, rotation=45)
plt.show()
```

### Observation 

The monthly sales trend shows that November and December had the highest sales, which may be related to increased shopping activity during the holiday season and Christmas. September also experienced relatively high sales, potentially due to increased demand at the beginning of the academic and business year.

## Sales by Sub-category

```python
subcat=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print(subcat)
```

```python
plt.figure(figsize=(12,6))
subcat.plot(kind="bar")
plt.title("Total Sales Over Subcategories")
plt.xlabel("Sub categories")
plt.ylabel("Total Sales($)")
plt.show()
```

### observation 

Phones and Chairs were the top-selling sub-categories, generating nearly the same total sales, with only about a $5K difference between them. Both significantly outperformed the remaining sub-categories, with a gap of over $100K compared to Storage, the third-highest category.

## Sales by Region

```python
regions=df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print(regions)
```

```python
plt.figure(figsize=(12,6))
regions.plot(kind="bar")
plt.title("Total Sales Over Regions")
plt.xlabel("Regions")
plt.ylabel("Total Sales($)")
plt.show()
```

### Observation 

The West region recorded the highest total sales, followed by the East region with a difference of approximately $41K. Central and South generated considerably lower sales, indicating that the majority of revenue was concentrated in the West and East regions.

## Top 10 States by Sales

```python
states=df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
print(states)
```

```python
plt.figure(figsize=(12,6))
states.plot(kind="bar")
plt.title("Total Sales Over Top 10 States")
plt.xlabel("States")
plt.ylabel("Total Sales($)")
plt.show()
```

### Observation 

California recorded the highest total sales at approximately $446K, outperforming New York by around $140K. In contrast, Virginia ranked tenth among the top selling states with approximately $71K in sales. This means California generated more than six times the sales of Virginia.

## Top 10 Products

```python
products=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
print(products)
```

```python
plt.figure(figsize=(12,6))
products.sort_values().plot(kind="barh")
plt.title("Total Sales Over Top 10 Products")
plt.ylabel("Products")
plt.xlabel("Total Sales($)")
plt.show()
```

### Observation

The Canon imageCLASS 2200 Advanced Copier ranked first with approximately $61.6K in sales, more than doubling the sales of the second-best-selling product. This indicates that a small number of products contributed disproportionately to total revenue.

## Top 10 Customers

```python
costumers=df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
print(costumers)
```

```python
plt.figure(figsize=(12,6))
costumers.plot(kind="bar")
plt.title("Top 10 Customers by Total Sales")
plt.xlabel("Customer name")
plt.ylabel("Total Sales($)")
plt.show()
```

### Observation

Sean Miller generated the highest total sales, reaching approximately $25K. Tamara Chand ranked second with around $19K in sales. The relatively small differences among the top customers suggest that sales were distributed across multiple high-value customers rather than relying heavily on a single customer.

# Key Insights

Technology was the highest-selling category.

Sales peaked in 2018, following steady growth after 2016.

November and December recorded the highest monthly sales, indicating a possible seasonal effect.

Phones and Chairs were the top-performing sub-categories.

The West region and California generated the highest total sales.

Sales were distributed across multiple high-value customers, with no single customer dominating overall sales.
